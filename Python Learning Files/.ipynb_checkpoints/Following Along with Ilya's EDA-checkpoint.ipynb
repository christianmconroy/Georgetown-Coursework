```python
interval = alt.selection_interval()

alt.Chart(iris).mark_point().encode(
    x='petalLength',
    y='petalWidth',
    color='species'
).properties(
    selection=interval
)
```

```python
interval = alt.selection_interval()

base = alt.Chart(iris).mark_point().encode(
    y='petalWidth',
    color=alt.condition(interval, 'species', alt.value('lightgray'))
).properties(
    selection=interval
)

base.encode(x='petalLength') | base.encode(x='sepalLength')
```

```python
departments = md.groupby('department_name')
rates = departments.mean().reset_index()
rates.head()
```

```python
interval = alt.selection_interval()

base = alt.Chart(rates).mark_point().encode(
    y='arrest_made',
    color=alt.condition(interval, 'department_name', alt.value('lightgray')),
    tooltip = 'department_name'
).properties(
    selection=interval
)

base.encode(x='subject_age') | base.encode(x='search_conducted')
```
